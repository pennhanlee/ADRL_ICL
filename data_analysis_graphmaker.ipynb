```python
manhattan_distances = {
    '50 Episodes Dataset': [6, 5, 4, 3, 2, 1],
    '100 Episodes Dataset': [6, 5, 4, 3, 2, 1],
    '200 Episodes Dataset': [6, 5, 4, 3, 2, 1],
    '400 Episodes Dataset': [6, 5, 4, 3, 2, 1]
}

paths_found = {
    '50 Episodes Dataset': [2, 6, 7, 12, 25, 46],
    '100 Episodes Dataset': [2, 7, 20, 24, 37, 50],
    '200 Episodes Dataset': [17, 17, 29, 32, 55, 51],
    '400 Episodes Dataset': [61, 61, 61, 61, 61, 61]
}

# Create a plot
plt.figure(figsize=(10, 6))

# Plot each pathfinder's data
for pathfinder in manhattan_distances.keys():
    plt.plot(manhattan_distances[pathfinder], paths_found[pathfinder], marker='o', label=pathfinder, linewidth=10)

# Add labels and title
plt.xlabel('Manhattan distance between attacker target state from original goal state', fontsize=15)
plt.ylabel('Paths from all other states to attacker target state', fontsize=15)
plt.tick_params(axis='both', which='major', labelsize=15)
plt.legend(title='Dataset Size', fontsize=15)
plt.grid(False)

# Save the plot as an image
plt.tight_layout()
plt.savefig('paths_vs_distance.png', dpi=300)

# Show the plot
plt.show()
```
